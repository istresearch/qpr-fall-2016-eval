```python
def file_grabber(init_path):
    out_files = []
    for i in os.listdir(init_path):
        if os.path.isdir(init_path + i):
            for e in os.listdir(init_path + i):
                out_files.append(init_path + i + '/' + e)
        else:
            out_files.append(init_path + i)
    return out_files
```
